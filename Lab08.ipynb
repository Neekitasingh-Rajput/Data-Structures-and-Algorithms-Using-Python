Q1. Create a Circular linked list with N nodes. All the inputs from user end.

Q2. Insert a new node in the linked list created in last question, at start and end

Q3. Insert a new node in the linked list created in last question, at

    a) after a specified node
    
    b) before a specified node
    
    c) fixed position

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CLinkedList:
    def __init__(self):
        self.head = None

    def display(self):
        if self.head is None:
            print("List is empty")
            return
        temp = self.head
        while True:
            print(temp.data, end=" -> ")
            temp = temp.next
            if temp == self.head:
                break
        print("(head)")

    def add_start(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            new_node.next = new_node
            return
        temp = self.head
        while temp.next != self.head:  
            temp = temp.next
        new_node.next = self.head
        self.head = new_node
        temp.next = self.head 

    def add_end(self, new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
            new_node.next = new_node
            return
        temp = self.head
        while temp.next != self.head:
            temp = temp.next
        temp.next = new_node
        new_node.next = self.head

    insert = add_end

    def add_pos(self, pos, new_data):
        if pos < 0:
            print("Invalid position!")
            return
        if pos == 0:
            self.add_start(new_data)
            return

        new_node = Node(new_data)
        temp = self.head
        i = 0

        while temp.next != self.head and i < pos - 1:
            temp = temp.next
            i += 1

        if temp.next == self.head and i < pos - 1:
            print("Position out of range!")
            return

        new_node.next = temp.next
        temp.next = new_node

    def add_b4_pos(self, pos, new_data):
        if pos <= 1:
            self.add_start(new_data)
            return
        new_node = Node(new_data)
        temp = self.head
        i = 0
        while temp.next != self.head and i < pos - 2:
            temp = temp.next
            i += 1

        if temp.next == self.head and i < pos - 2:
            print("Position out of range!")
            return

        new_node.next = temp.next
        temp.next = new_node

    def add_after_pos(self, pos, new_data):
        new_node = Node(new_data)
        temp = self.head
        i = 0
        while temp.next != self.head and i < pos:
            temp = temp.next
            i += 1
        if temp.next == self.head and i < pos:
            print("Position out of range!")
            return
        new_node.next = temp.next
        temp.next = new_node

CLL = CLinkedList()
n = int(input("Enter the number of nodes you want to enter: "))
for i in range(n):
    data = input(f"Enter the data for node {i+1}: ")
    CLL.insert(data)

print("\nOriginal Circular Linked List:")
CLL.display()

print("\nAfter adding 0 at the start:")
CLL.add_start(0)
CLL.display()

print("\nAfter adding 6 at the end:")
CLL.add_end(6)
CLL.display()

print("\nAfter adding 2.5 before position 3:")
CLL.add_b4_pos(3, 2.5)
CLL.display()

print("\nAfter adding 3.5 after position 4:")
CLL.add_after_pos(4, 3.5)
CLL.display()

print("\nAfter adding 4.5 at position 7:")
CLL.add_pos(7, 4.5)
CLL.display()
```

```python
#1. Write a program that prints numbers from 1 to 10 using an infinite loop.
#All numbers should get printed in the same line.

i = 1
while(1):
    print(i, end=' ')
    i += 1
    if i>10:
        break
```

```python
#2. Write a program that prints all unique combinations of 1, 2 and 3.

numbers = [1,2,3]
for i in numbers:
    for j in numbers:
        if j!=i:
            for k in numbers:
                if k!=i and k!= j:
                    print([i,j,k])
```

```python
#3. Write a program to print all prime numbers from 1 to 300.

for i in range(3,301):
    is_prime = True
    for j in range(2,i):
        if i%j == 0:
            is_prime = False
    if is_prime:
        print(i ,end=' ')
```

```python
#4. Write a program to count the number of alphabets and number of digits in the string ‘CHRIST-560029’.

string = 'CHRIST-560029'
alphabet_count = 0
digit_count = 0
for char in string:
    if char.isalpha():
        alphabet_count += 1
    if char.isdigit():
        digit_count += 1
print("No. of alphabets: ",alphabet_count)
print("No. of digits: ",digit_count)
```

```python
#5. Suppose there are four flag variables w,x,y,z, write a program to check in multiple ways whether one of them is true.

# Define the flag variables
w = False
x = False
y = True
z = False

# Method 1: Using `or` operator
if w or x or y or z:
    print("At least one of the flags is True (using `or`).")
else:
    print("None of the flags are True (using `or`).")

# Method 2: Using `any()` function
if any([w, x, y, z]):
    print("At least one of the flags is True (using `any()`).")
else:
    print("None of the flags are True (using `any()`).")

# Method 3: Using a loop
flag_found = False
for flag in [w, x, y, z]:
    if flag:
        flag_found = True
        break
if flag_found:
    print("At least one of the flags is True (using a loop).")
else:
    print("None of the flags are True (using a loop).")

# Method 4: Using a generator expression
if any(flag for flag in [w, x, y, z]):
    print("At least one of the flags is True (using a generator expression).")
else:
    print("None of the flags are True (using a generator expression).")
```

```python
#7. If the three sides of a triangle are entered thorough the keyboard , write a program to check
# whether the triangle is isosceles , equilateral , scalene or right angled triangle.

a = int(input("Enter length of side 1: "))
b = int(input("Enter length of side 2: "))
c = int(input("Enter length of side 3: "))

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("The triangle is equilateral.")
    elif a == b or a == c or b == c:
        print("The triangle is isosceles.")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("The triangle is a right-angled triangle.")
    else:
        print("The triangle is scalene.")
else:
    print("The given sides do not form a valid triangle.")
```

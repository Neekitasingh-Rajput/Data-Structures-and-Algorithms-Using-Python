```python
#A.
#Create a linked list with N nodes. All the inputs from user end. 
#Display the list with and arrow mark indicating Linked List.

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert(self,new_data):
        new_node = Node(new_data)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = new_node

    def display(self):
        if self.head is None:
            print("Empty list")
        else:
            link_str = 'Linked list: '
            temp = self.head
            while temp:
                link_str += str(temp.data) + " -> "
                temp = temp.next
            link_str += "None"
            print(link_str)

    def remove_start(self):
        self.head = self.head.next

    def remove_after_pos(self,pos):
        count = 0
        temp = self.head
        while temp.next != None and count < pos:
            temp = temp.next
            count += 1
        if temp.next is None or pos < 0:
            print("Invalid position!")
        else:
            temp.next = temp.next.next

    def remove_b4_pos(self,pos):
        count = 0
        temp = self.head
        while temp.next != None and count < pos-2:
            temp = temp.next
            count += 1
        if temp.next is None or pos <= 0:
            print("Invalid position!")
        else:
            temp.next = temp.next.next

    def remove_pos(self,pos):
        if pos == 0:
            self.remove_start()
        count = 0
        temp = self.head
        while temp.next != None and count < pos-1:
            temp = temp.next
            count += 1
        if temp.next is None or pos < 0:
            print("Invalid position!")
        else:
            temp.next = temp.next.next
        
LL= LinkedList()
n = int(input("Enter the no. of nodes:"))
for i in range(n):
    node = input(f"Enter the data for node {i+1}:")
    LL.insert(node)
    
print()    
LL.display()
```

```python
#C.
#Remove the nodes in the linked list created in last question, for the following cases given by user-
#a. After a specified node
#b. Before a specified node
#c. From a fixed position.
#Display the list with and arrow mark indicating Linked List for all the questions.


print("#Original")
LL.display()
print("#Remove node after pos 3 (elem : 6)")
LL.remove_after_pos(3)
LL.display()
print("#Remove node before pos 5 (elem : 7)")
LL.remove_b4_pos(5)
LL.display()
print("#Remove node at pos 4 (elem : 8)")
LL.remove_pos(4)
LL.display()
```

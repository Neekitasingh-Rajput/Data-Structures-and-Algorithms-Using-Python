```python
#A. Implement Stack Using Normal Array.

import array
arr = array.array('i')

def stack_push(data,arr):
    arr.insert(0,data)
    
def stack_pop(arr):
    arr.pop(0)

stack_push(10,arr)
print(arr," #Added 10\n")
stack_push(20,arr)
print(arr, "  #Added 20\n")      
stack_push(30,arr)
print(arr, "  #Added 30\n")
stack_pop(arr)
print(arr, "  #Removed 30\n")
stack_pop(arr)
print(arr, "  #Removed 20\n")
stack_push(15,arr)
print(arr, "  #Added 15\n")
```

```python
#B. Implement Stack Using Linked List.

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class Stack:
    def __init__(self):
        self.head = None

    def display(self):
        temp = self.head
        print("Stack:", end = ' ')
        while temp is not None:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None",end = "   ")

    def stack_push(self, newdata):
        new_node = Node(newdata)
        new_node.next = self.head
        self.head = new_node

    def stack_pop(self):
        if self.head is None:
            print("Stack Underflow! Cannot pop from an empty stack.")
            return
        self.head = self.head.next
        
S = Stack()
n = int(input("Enter no. of elements to push:"))
for i in range(n):
    elem = int(input("\nEnter the element to push:"))
    S.stack_push(elem)
    S.display()
    print(f"# Pushed {elem}")

print("\n#Pop")
S.stack_pop()
S.display()
print("# Popped 40\n")
S.stack_pop()
S.display()
print("# Popped 30")
```

```python
#C. Convert an Infix notation to its equivalent Postfix notation using Stack.

def precedence(op):
    if op == '+' or op == '-':
        return 1
    if op == '*' or op == '/':
        return 2
    if op == '^':
        return 3
    return 0

def infix_to_postfix(expression):
    stack = []
    postfix = ""

    for char in expression:
        if char.isalnum():
            postfix += char
        elif char == '(':
            stack.append(char)
        elif char == ')':
            while stack and stack[-1] != '(':
                postfix += stack.pop()
            stack.pop()
        else:
            while stack and precedence(stack[-1]) >= precedence(char):
                postfix += stack.pop()
            stack.append(char)

    while stack:
        postfix += stack.pop()

    return postfix

expression = input("\nEnter infix expression: ")
print("Postfix Expression:", infix_to_postfix(expression))

expression = input("\nEnter infix expression: ")
print("Postfix Expression:", infix_to_postfix(expression))

expression = input("\nEnter infix expression: ")
print("Postfix Expression:", infix_to_postfix(expression))
```

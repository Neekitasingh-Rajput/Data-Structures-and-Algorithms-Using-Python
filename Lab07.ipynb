```python
class DNode:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None

class DLinkedList:
    def __init__(self):
        self.head = None
        self.last = None

    def traverse_forw(self):
        if self.head is None:
            print("List is empty!")
            return
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

    def traverse_back(self):
        if self.head is None:
            print("List is empty!")
        temp = self.last
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.prev
        print("None")
    
    def add_start(self,new_data):
        new_node = DNode(new_data)
        if self.head == None:
            self.head = self.last = new_node
            return
        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node

    def add_end(self,new_data):
        new_node = DNode(new_data)
        if self.head == None:
            self.head = self.last = new_node
            return
        self.last.next = new_node
        new_node.prev = self.last
        self.last = new_node

    insert = add_end

    def add_after_pos(self,pos,new_data):
        if pos == 0:
            self.add_start(new_data)
            return
        i = 0
        temp = self.head
        while temp and i < pos:
            temp = temp.next
            i += 1
        if temp is None:
            print("Position out of range!")
            return
            
        new_node = DNode(new_data)
        new_node.next = temp.next
        new_node.prev = temp

        if temp.next != None:
            temp.next.prev = new_node
        temp.next = new_node

        if new_node.next is None:
            self.last = new_node
            
    def add_b4_pos(self,pos,new_data):
        if pos == 0:
            self.add_start(new_data)
            return
        i = 0
        temp = self.head
        while temp and i < pos-2:
            temp = temp.next
            i += 1
        if temp is None:
            print("Position out of range!")
            return
            
        new_node = DNode(new_data)
        new_node.next = temp.next
        new_node.prev = temp

        if temp.next != None:
            temp.next.prev = new_node
        temp.next = new_node

        if new_node.next is None:
            self.last = new_node
            
    def add_pos(self,pos,new_data):
        if pos == 0:
            self.add_start(new_data)
            return
        i = 0
        temp = self.head
        while temp and i < pos-1:
            temp = temp.next
            i += 1
        if temp is None:
            print("Position out of range!")
            return
            
        new_node = DNode(new_data)
        new_node.next = temp.next
        new_node.prev = temp

        if temp.next != None:
            temp.next.prev = new_node
        temp.next = new_node

        if new_node.next is None:
            self.last = new_node
```

```python
#A. Create a doubly linked list with N nodes. All the inputs from user end.

dll = DLinkedList()
a = int(input("Enter the number of nodes: "))
for i in range(a):
    data = input(f"Enter data for node {i+1}: ")
    dll.insert(data)

print("\nDoubly Linked List (Forward):")
dll.traverse_forw()

print("\nDoubly Linked List (Backward):")
dll.traverse_back()
```

```python
#B. Insert a new node in the linked list created in last question, at start and end.

print("Original List:")
dll.traverse_forw()

print("\nAfter adding 0 at the start:")
dll.add_start(0)
dll.traverse_forw()

print("\nAfter adding 6 at the end:")
dll.add_end(6)
dll.traverse_forw()
```

```python
#C. Insert a new node in the linked list created in last question, at
#i.after a specified pos
#ii.before a specified pos
#iii.fixed pos

print("\nAfter adding 2.5 at position 3:")
dll.add_pos(3,2.5)
dll.traverse_forw()

print("\nAfter adding 0.5 b4 position 1:")
dll.add_b4_pos(1,0.5)
dll.traverse_forw()

print("\nAfter adding 4.5 after position 6:")
dll.add_after_pos(6,4.5)
dll.traverse_forw()
```

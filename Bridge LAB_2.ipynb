```python
#1. Perform the following operations on a list of names:

#a. Create a list of 5 names –'Anil', 'Amol, ‘Aditya’,’Avi’,’Alka’

nlist = ['Anil','Amol','Aditya','Avi','Alka']
print(nlist, "#List created")
        
#b. Insert a name ‘Anuj’ before ‘Aditya’

nlist.insert(2,'Anuj')
print(nlist, "#'Anuj' inserted before 'Aditya'.")

#c. Append a name ‘Akhil’.

nlist.append('Akhil')
print(nlist, "#'Akhil' appended in the list.")

#d. Delete ‘Avi’ from the list

nlist.remove('Avi')
print(nlist, "#'Avi' deleted.")

#e. Replace ‘Anil’ with ‘Anilkumar’

nlist[0] = 'Anilkumar'
print(nlist, "#'Anil' replaced with 'Anilkumar'.")

#f. Sort all the names in the list

nlist.sort()
print(nlist, "#List sorted.")

#g. Print reversed sorted list

nlist.reverse()
print(nlist, "#List reverse sorted.")
```

```python
#2. Perform the following operations on a list of number:

#a. Create a list of 5 odd numbers

odd_num = [1, 3, 5, 7, 9]
print("Odd_List:", odd_num)

#b. Create a list of 5 even numbers

even_num = [2, 4, 6, 8, 10]
print("Even_List:", even_num)

#c. Combine the two lists.

combined_list = odd_num + even_num
print("Combined_List:", combined_list)

#d. Add prime numbers 11,17,29 at the beginning of the combined list.

primes = [11,17,29]
combined_list = primes + combined_list
print("List with primes added", combined_list)

#e. Report how many elements are present in the list.

print("Number of elements in the list:", len(combined_list))

#f. Replace last 3 numbers in the list with 100,200,300.

combined_list[-3: ] = [100,200,300]
print("Last three elements changed" , combined_list)
      
#g. Delete all the numbers in the list.

combined_list.clear()
print("List with all the numbers removed:" , combined_list)

#h. Delete the list.

del(combined_list)

try:
    print(combined_list)
except NameError:
    print("The list has been deleted.")
```

```python
#3. Write a program to generate and store in a list 20 random numbers in the range 10 to 100.
#From the list delete all those entries which have value between 20 and 50.
#Print the remaining list.


import random

random_numbers = [random.randint(10, 100) for x in range(20)]
print("Original List of Random Numbers:", random_numbers)

remaining_numbers = [num for num in random_numbers if num not in range(20,50)]
print("Remaining List after Deleting Numbers between 20 and 50:", remaining_numbers)
```

```python
#4. Write a program to add two 3x4 matrices.

#for taking matrices input.
def input_matrix(rows,cols,matrix_name):
    print(f" Enter the elements for {matrix_name}({rows}x{cols}):")
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = int(input(f" Enter the value for element [{i+1}][{j+1}]:"))
            row.append(element)
        matrix.append(row)
    return matrix

#for adding the matrices.
def add_matrices(matrix1, matrix2):
    rows = len(matrix1)
    cols = len(matrix1[0])
    addition = [[matrix1[i][j] + matrix2[i][j] for j in range(cols)] for i in range(rows)]

    return addition

rows = int(input("Enter the no. of rows:"))
cols = int(input("Enter the no. of columns:"))

matrix1 = input_matrix(rows,cols, "Matrix1")
matrix2 = input_matrix(rows,cols, "Matrix2")

result = add_matrices(matrix1, matrix2)

print("\nMatrix 1:")
for row in matrix1:
    print(row)

print("\nMatrix 2:")
for row in matrix2:
    print(row)

print("\nSum of the two matrices:")
for row in result:
    print(row)
```

```python
#5. Suppose a list contains several words. Write a program to create another list
#that contains the first character of each word present in the first list.

my_list = ['Apple', 'Banana', 'Chikoo', 'Guava', 'Pineapple']
first_alpha_list = []

for i in my_list:
    first_alpha = i[0]
    first_alpha_list.append(first_alpha)

print("List:", my_list)
print("List of first characters:", first_alpha_list)
```

```python
#6. A list contains names of boys and girls as its elements.
#Boys’ names are stored as tuples.
#Write a python program to find out number of boys and girls in the list.

names_list = [('John',), 'Emily', ('Mike',), 'Sophia', ('Adam',), 'Olivia']

boys_count = 0
girls_count = 0

for name in names_list:
    if type(name) == tuple:
        boys_count += 1
    else:
        girls_count += 1

print("Number of boys:", boys_count)
print("Number of girls:", girls_count)
```

```python
#7. Given the following tuple
#(‘F’, ’l’,’ a’, ’b’, ’b’, ’e’, ’r’, ’g’, ’a’, ’s’, ’t’, ’e’, ’d’)

#Write a python program to carry out the following operations:

my_tuple = ('F', 'l', 'a', 'b', 'e', 'r', 'g', 'a', 's', 't', 'e', 'd')

#a. Add an ! at the end of the tuple.

print("Tuple:", my_tuple)
my_tuple = my_tuple + ('!',)
print("Edited Tuple:" , my_tuple)

#b. Convert a tuple to a string.

string = ''.join(my_tuple)
print("Converted string:", string)

#c. Extract (‘b’,’b’) from the tuple.

extracted_tuple = tuple(element for element in my_tuple if element == 'b')
print("Extracted tuple:", extracted_tuple)

# d. Find out the number of occurrences of 'e' in the tuple

e_count = my_tuple.count('e')
print(f"Number of occurrences of 'e': {e_count}")

# e. Check whether 'r' exists in the tuple

r_exists = 'r' in my_tuple
print(f"Does 'r' exist in the tuple? {'Yes' if r_exists else 'No'}")

# f. Convert the tuple to a list

my_list = list(my_tuple)
print("Converted list:", my_list)

# g. Delete characters 'b', 'b', 'e', 'r' from the tuple

filtered_list = [element for element in my_tuple if element not in ('b', 'e', 'r')]
new_tuple = tuple(filtered_list)
print("Modified tuple after deletion:", new_tuple)
```

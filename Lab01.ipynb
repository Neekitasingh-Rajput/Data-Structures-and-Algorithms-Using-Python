```python
'''
B) Write a program to create a class called "Dog" with a name and breed attribute.
Create two instances of the "Dog" class, set their attributes using the constructor and 
modify the attributes using the setter methods and print the updated values.
'''

class Dog():

    def __init__(self,name,breed):
        self.name = name
        self.breed = breed

    def set_name(self,name):
        self.name = name

    def set_breed(self,breed):
        self.breed = breed

    def display(self):
        print(f"Name: {self.name}, Breed: {self.breed}")

dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Lucy", "Labrador")

dog1.display()
dog2.display()

dog1.set_name("Max")
dog1.set_breed("German Shepherd")

dog2.set_name("Bella")
dog2.set_breed("Beagle")

dog1.display()
dog2.display()
```

```python
#2. 

'''
Create a class with multiple parameters with different arguments. Print arguments
with proper messages for all the following cases:

i. A single number
ii. List of multiple numbers
iii. Single string
iv. Multiple strings
v. List of numbers and strings
'''

class ArgumentHandler:

    def __init__(self,arg):
        self.arg = arg

    def display(self):
        if type(self.arg) == int:
            print(f"Single Number: {self.arg}")
        elif type(self.arg) == str:
            print(f"Single String: '{self.arg}'")
        elif type(self.arg) == list:
            if all(type(x) == int for x in self.arg):
                print(f"List of Numbers: {self.arg}")
            elif all(type(x) == str for x in self.arg): 
                print(f"List of Strings: {self.arg}")
            else:
                print(f"List of Mixed Types: {self.arg}")
        else:
            print("Unsupported type!")
            
handler1 = ArgumentHandler(42)
handler1.display()

handler2 = ArgumentHandler([1, 2, 3, 4, 5])
handler2.display()

handler3 = ArgumentHandler("Hello")
handler3.display()

handler4 = ArgumentHandler(["First", "Second", "Third"])
handler4.display()

handler5 = ArgumentHandler(["Alice", 3, "Bob", 7])
handler5.display()
```

```python
#3
'''
Create a flashcard game like following using python class (Uses interactive messages).

i. Take an word and its meaning as input from the user.
ii. Create a class named flashcard, use the __init__() function to assign values
for Word and Meaning.
iii. Now we use the __str__() function to return a string that contains the word
and meaning.
iv. Store the returned strings in a list named flash.
v. Use a while loop to print all the stored flashcards.
'''

class Flashcard:
    def __init__(self, word, meaning):
        self.word = word
        self.meaning = meaning

    def __str__(self):
        return f"Word: {self.word}, Meaning: {self.meaning}"

flash = []

while True:
    word = input("Enter a word (or type 'stop' to finish): ")
    if word.lower() == 'stop':
        break
    meaning = input(f"Enter the meaning of '{word}': ")
    flashcard = Flashcard(word, meaning)
    flash.append(str(flashcard))

print("\nYour Flashcards:")
for card in flash:
    print(card)
```

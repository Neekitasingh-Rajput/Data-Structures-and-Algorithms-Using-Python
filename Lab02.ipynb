```python
#1. Create the following array and display them using python
'''
a) Array of integers
b) Array of floating point numbers
c) Array of alphabets
d) Array of strings
'''

import array

# a) Array of integers
int_array = array.array('i', [1, 2, 3, 4, 5])
print("Array of integers:", int_array)

# b) Array of floating point numbers
float_array = array.array('f', [1.1, 2.2, 3.3, 4.4])
print("Array of floating point numbers:", float_array)

# c) Array of alphabets
# Using a list for alphabets since `array` doesn't support characters
alphabets_array = ['a', 'b', 'c', 'd', 'e']
print("Array of alphabets:", alphabets_array)

# d) Array of strings
# Using a list for strings as well since `array` supports only numbers
strings_array = ["apple", "banana", "cherry"]
print("Array of strings:", strings_array)
```

```python
#2. Perform the following operations on the above arrays
'''
a) Find length of the array
b) Find location of some element
c) Add some elements into the array at –
    i) Start ii) End iii) before some element iv) after some element.
d) Delete some elements from the array at –
    i) Start ii) End iii) before some element iv) after some element.
e) Update some values from the array
'''

from array import array

arr = array('i', [10, 20, 30, 40, 50]) 
print("Original Array:", arr)

print("Length:", len(arr))# Length

# Location
print("Location of 30:", arr.index(30) if 30 in arr else "Not found")

# Add elements
arr.insert(0, 5) # Start
arr.append(60) # End
arr.insert(arr.index(40), 35) # Before 40
arr.insert(arr.index(20) + 1, 25) # After 20

print("After Additions:", arr)

#Delete elements
arr.pop(0) # Start
arr.pop() # End
arr.pop(arr.index(50) - 1) # Before 50
arr.pop(arr.index(20) + 1) # After 20

print("After Deletions:", arr)

arr[2] = 100
print("After Update:", arr)
```

```python
#3. Perform the following operation on the above array:
'''
f) Reverse the array
g) Sort the array
h) Count the number of existence of a specific element in the array.
'''

from array import array
arr = array('i', [10, 20, 30, 40, 50, 20, 30, 20])
print("Original Array:", arr)

arr.reverse()
print("Reversed Array:", arr)

arr = array('i', sorted(arr))
print("Sorted Array:", arr)

element = 20
count = arr.count(element)
print(f"Count of {element}:", count)
```

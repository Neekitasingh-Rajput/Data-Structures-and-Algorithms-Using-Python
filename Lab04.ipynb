```python
#Implement Merge sort on two list of given integers.

list1 = [12,43,-18,65,76,66,87]
list2 = [100,45,45,96,87,23,17]

def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])
    
    return merge(left_half, right_half)

def merge(left, right):
    sorted_array = []
    i = j = 0
    
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_array.append(left[i])
            i += 1
        else:
            sorted_array.append(right[j])
            j += 1
    
    while i < len(left):
        sorted_array.append(left[i])
        i += 1
    
    while j < len(right):
        sorted_array.append(right[j])
        j += 1
    
    return sorted_array


my_list = list1 + list2
sorted_arr = merge_sort(my_list)
print(f"List1 = {list1} \nList2 = {list2}")
print("Sorted merged array (merge sort):", sorted_arr)
```

```python
#Write a python program to accept input numbers from user until they enter a 0. 
#Store them in an array. Sort them using Insertion sort.

import array
arr = []
while True:
    elem = int(input(f"Enter the elem you want to add to the array(or enter 0 to stop):"))
    if elem == 0:
        break
    else:
        arr.append(elem)
print("\nArray:", arr)

def insertion_sort(arr):
    for current in range(1,len(arr)):
        temp = arr[current]
        comp = current - 1
        while comp >= 0 and temp < arr[comp]:
                arr[comp + 1] = arr[comp]
                comp -= 1
        arr[comp+1] = temp
    return arr

print("Sorted array (insertion sort):", insertion_sort(arr))
```

```python
#Implement Quick sort on a list of given integers.

def quick_sort(arr, low, high):
    if low < high:
        pivot_index = partition(arr, low, high)
        
        quick_sort(arr, low, pivot_index - 1)
        quick_sort(arr, pivot_index + 1, high)

def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    return i + 1

arr = [10, 7, 8, 9, 1, 5]
print("Array:", arr)
quick_sort(arr, 0, len(arr) - 1)
print("Sorted array (quick sort):",arr)
```

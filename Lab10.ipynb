```python
#B. Implement Queue Using Linked List.

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.head = None

    def enqueue(self, newdata):
        new_node = Node(newdata)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = new_node

    def dequeue(self):
        if self.head == None:
            print("Queue Underflow!")
            return
        print(f"# Dequeued {self.head.data}")
        self.head = self.head.next

    def display(self):
        temp = self.head
        print("Queue:", end = ' ')
        while temp is not None:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")
        
Q = Queue()
n = int(input("Enter no. of elements to add to the queue:"))
for i in range(n):
    elem = int(input("\nEnter the element to enqueue:"))
    Q.enqueue(elem)
    Q.display()
    print(f"# Enqueued {elem}")

print("\n#Dequeuing")
print()
Q.dequeue()
Q.display()
print()
Q.dequeue()
Q.display()
```

```python
#C. Write a program to check if two queues are equal or not.

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class Queue:
    def __init__(self):
        self.head = None

    def enqueue(self, newdata):
        new_node = Node(newdata)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next is not None:
            temp = temp.next
        temp.next = new_node
        
    def is_equal(self, other):
        temp1 = self.head
        temp2 = other.head

        while temp1 and temp2:
            if temp1.data != temp2.data:
                return False
            temp1 = temp1.next
            temp2 = temp2.next

        return temp1 is None and temp2 is None

    def display(self):
        temp = self.head
        print("Queue:", end = ' ')
        while temp is not None:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

#taking queue inputs:-
Q1 = Queue()
n1 = int(input("Enter no. of elements to add to Q1:"))
for i in range(n1):
    elem = int(input("Enter the element to enqueue:"))
    Q1.enqueue(elem)

Q2 = Queue()
n2 = int(input("\nEnter no. of elements to add to Q2:"))
for i in range(n2):
    elem = int(input("Enter the element to enqueue:"))
    Q2.enqueue(elem)

#to check if Q1 == Q2:
print("\n#Q1",end=" ")
Q1.display()
print("#Q2",end=" ")
Q2.display()

if Q1.is_equal(Q2):
    print("\nThe two queues are EQUAL!")
else:
    print("\nThe two queues are NOT equal!")
```

```python
#1.
#Implement selection sort for a list of numbers.

import array
arr = array.array('i', [100, 30, 12, 35, -3, 0])

#Selection sort

def selection_sort(arr):
    for i in range(len(arr)):
        min = arr[i]
        pos = i
        for j in range(i, len(arr)):
            if arr[i] > arr[j]:
                min = arr[j]
                pos = j
                arr[i] , arr[j] = arr[j], arr[i]
    return arr

print("Sorted elements (Selection sort):", selection_sort(arr))
```

```python
#2.
#Write a python program to accept input numbers from users until they enter 0.
#Store them in an array. Sort them using Bubble sort.

import array
arr = []
while True:
    elem = int(input(f"Enter the elem you want to add to the array(or enter 0 to stop):"))
    if elem == 0:
        break
    else:
        arr.append(elem)
print("Array:", arr)
#Bubble sort

def bubble_sort(arr):
    for i in range(len(arr)):
        for j in range(0,len(arr)-1-i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr
                
print("Sorted Array (Bubble sort):", bubble_sort(arr))
```

```python
#3.
#Write a python program to accepts names from user and sort them in ascending order.

names = []
n = int(input("Enter the no. of names you want to add to the list"))
for i in range(n):
    name = str(input(f"Enter name {i+1}:"))
    names.append(name)
print("Names:", names)

def bubble_sort(list):
    for i in range(len(list)):
        for j in range(0,len(list)-1-i):
            if list[j] > list[j+1]:
                list[j], list[j+1] = list[j+1], list[j]
    return list

print("Names arranged in ascending order:", bubble_sort(names))
```
